## Method #1: isin funtion/method

```python
# df1.apply(tuple, 1).isin(df2.apply(tuple, 1))
result1 = df1[~df1.apply(tuple, 1).isin(df2.apply(tuple, 1))]
result1.head()
```

```python
result2 = df2[~df2.apply(tuple, 1).isin(df1.apply(tuple, 1))]
# print(result1)
result2.head()
```

## Method #2 - Merge

```python
# df1.merge(df2, indicator=True)
# df1.merge(df2, indicator=True, how='outer') # Look for the "_merge" column
result2=df1.merge(df2, indicator=True, how='outer').loc[lambda v: v['_merge'] != 'both'] # 'both' NOT in both df's
```
